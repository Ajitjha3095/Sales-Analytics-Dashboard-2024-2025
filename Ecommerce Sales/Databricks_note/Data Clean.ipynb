```sql
SELECT COUNT(*) AS total_rows FROM ecommerceJune2026.sales.ecommerce_sales;
```

```sql
SELECT
    SUM(CASE WHEN Order_ID IS NULL THEN 1 ELSE 0 END) AS null_order_id,
    SUM(CASE WHEN Product_Name IS NULL THEN 1 ELSE 0 END) AS null_product,
    SUM(CASE WHEN Order_Date IS NULL THEN 1 ELSE 0 END) AS null_date,
    SUM(CASE WHEN Quantity IS NULL THEN 1 ELSE 0 END) AS null_quantity
FROM ecommerceJune2026.sales.ecommerce_sales;
```

```sql
SELECT COUNT(*) AS invalid_sales
FROM ecommerceJune2026.sales.ecommerce_sales
WHERE Total_Sales <> Quantity * Unit_Price;
```

```sql
CREATE OR REPLACE VIEW ecommerce.sales.vw_sales AS

SELECT
    Order_ID,
    Order_Date,
    YEAR(Order_Date) AS Year,
    MONTH(Order_Date) AS Month,
    QUARTER(Order_Date) AS Quarter,
    Product_Name,
    Category,
    Region,
    Customer_Type,
    Payment_Method,
    Quantity,
    Unit_Price,
    Total_Sales
FROM ecommerceJune2026.sales.ecommerce_sales;
```
